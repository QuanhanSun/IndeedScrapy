```python
import json
import re
```

```python
with open('data_scientist.json', 'r') as f:
    data = json.load(f)
```

```python
res = []
for i in range(len(data)):
    temp = re.findall(r" \(e\)(.+)                try ", data[str(i)]['posting'].replace('\n',''))[0]
    temp1 = re.findall(r"\(e\) \{\};(.+)try", temp)[0].strip()
    res.append(temp1)
```
